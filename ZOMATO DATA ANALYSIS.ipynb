```python
import pandas as pd
df=pd.read_csv('zomato_data.csv')
print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# To count the type of restaurant
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Type of Restaurant')
```

```python
# To count the number of words for the type of restaurant
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="Blue", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

```python
# To print restaurant with maximum votes
max_votes = df['votes'].max()
restaurant_with_max_votes = df.loc[df['votes'] == max_votes, 'name']
print('Restaurant with max votes:', restaurant_with_max_votes)
```

```python
# To count how many are intrested to order online
sns.countplot(x=df['online_order'])
```

```python
# To measure the rate distribution
plt.hist(df['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()
# The below graph shows that the rating is between 3.5 and 4
```

```python
# The approx cost for 2 people
couple_cost = df['approx_cost(for two people)']
sns.countplot(x=couple_cost)
```
